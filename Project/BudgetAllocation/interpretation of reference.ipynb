```python
target_col = '사업명'
ministry_col = '부처명'

# 찾고자 하는 키워드 및 부처명 입력
keyword = 'BK|두뇌|플러스'
target_ministry = '교육부'

# ---------------------------------------------------------
# [조건 1] 사업명에 키워드가 포함된 행 (기존)
mask_keyword = df_사업과제[target_col].str.contains(keyword, na=False, case=False)

# [조건 2] 부처명이 특정 부처와 정확히 일치하는 행
# (만약 부처명도 '포함'되는 것으로 찾고 싶다면 mask_keyword처럼 str.contains를 쓰셔도 됩니다)
mask_ministry = df_사업과제[ministry_col] == target_ministry

# 1. 두 조건을 '&' (AND) 연산자로 결합합니다. (반드시 양쪽 조건에 괄호 씌우기!)
final_mask = (mask_keyword) & (mask_ministry)
# ---------------------------------------------------------

# 2. 결합된 조건(final_mask)으로 필터링한 후 '고유값' 추출
keyword_list = df_사업과제.loc[final_mask, target_col].unique().tolist()

# [Hook] 눈으로 직접 결과 확인하기
print(f"✅ [{target_ministry}] 소속이면서 '{keyword}' 키워드를 포함하는 고유 사업 수: {len(keyword_list)}개")
print("-" * 50)

# 세로로 보기 좋게 출력 (상위 20개만)
for i, val in enumerate(keyword_list[:20], 1):
    print(f"{i}. {val}")

if len(keyword_list) > 20:
    print("... (이하 생략)")
```

됐다. 이제 이 변수들만 포함을 시키면 된다. BK21이 2015년에만 2번 이름으로 불렸던 것이다. 5개를 모두 포함하는 게 맞는 거다.

```python
# '사업명' 컬럼의 값들이 keyword_list 안에 포함되어 있는지(isin) 확인 후 개수 합산
target_count = df_사업과제['사업명'].isin(keyword_list).sum()

print(f"해당 샘플의 개수: {target_count:,}개")
```

```python
# 과제수행연도 컬럼이 텍스트(String)로 되어있을 수 있으므로, 안전하게 숫자형(int)으로 먼저 변환합니다.
# (이미 숫자형이라면 이 줄은 생략해도 되지만, 확실히 해두는 것이 좋습니다)
df_사업과제['과제수행연도'] = df_사업과제['과제수행연도'].astype(int)

# ---------------------------------------------------------
# [조건 1] '사업명'이 keyword_list에 포함되는 데이터
mask_keyword = df_사업과제['사업명'].isin(keyword_list)

# [조건 2] '과제수행연도'가 2013년 이상, 2016년 이하인 데이터
mask_year = df_사업과제['과제수행연도'].between(2013, 2016)
# ---------------------------------------------------------

# 두 조건을 '&' (AND) 로 묶어서 최 필터링을 수행하고 새로운 df로 정의합니다.
# (.copy()를 붙여 원본과 완전히 분리된 안전한 데이터프레임으로 만듭니다)
df_2 = df_사업과제[(mask_keyword) & (mask_year)].copy()

# [Hook] 데이터가 잘 추출되었는지 확인
print(f"✅ 조건 만족 데이터 개수: {len(df_2):,}개")
print("-" * 40)
print(df_2['과제수행연도'].value_counts().sort_index()) # 연도별 분포 확인

df_2.head(3)
```

매년 예산이 새로 편성된다는 이유로 같은 과제임에도 해마다 새로운 과제고유번호를 발급하고, 정작 족보를 이어야 할 이전과제고유번호는 입력자가 제대로 안 적어서 결측치가 투성이인 것. 이것이 대한민국 NTIS 데이터의 가장 큰 고질병입니다.

이 난관 앞에서 "Sliding Window로 계보를 추적해야 하나?" 하고 고민하시는 것은 데이터 분석가로서 너무나 당연하고 훌륭한 접근입니다. 하지만 결론부터 말씀드리면, 선생님께서 과거에 이미 완벽한 정답을 내리셨습니다.

선생님께서 올려주신 과거 메모(reverse engineering(0421))를 보면 이런 구절이 있습니다.

"사실 나는 이전과제고유번호 변수를 아예 버릴 생각이야... (중략)... 이번 연구에서는 시차는 무시하는 게 좋을 것 같아."

이 과거의 결단이 왜 '신의 한 수'인지, 그리고 선행연구(458개 샘플)는 도대체 이 문제를 어떻게 피해 갔는지 데이터 관점에서 역설계해 드립니다.

 선행연구는 어떻게 이 문제를 회피했을까? (추론)
이전과제고유번호 결측치가 이토록 많다면, 선행연구의 저자들(대학 연구진) 역시 이 변수를 이용해 다년도 과제의 계보(Lineage)를 완벽하게 추적하는 것은 불가능에 가깝다는 것을 깨달았을 것입니다. 저자들은 십중팔구 다음 두 가지 우회로 중 하나를 택했습니다.

1. "과제명(텍스트)" 또는 "PI(연구책임자)"를 새로운 Key로 사용
번호는 매년 시스템이 기계적으로 바꾸지만, **연구개발과제명(국문)**과 **연구책임자**의 이름, 그리고 소속은 다년도 과제 내내 거의 바뀌지 않습니다.
저자들은 과제고유번호를 버리고, "과제명이 같고 책임자가 같은 데이터"를 Groupby로 묶어서 하나의 과제로 병합(Sum/Mean)했을 확률이 매우 높습니다.

2. 각 연도를 독립된 스냅샷(Snapshot)으로 취급
1년 차 예산이 투입된 2013년의 과제 A와, 2년 차 예산이 투입된 2014년의 과제 A'를 연결하지 않고, 그냥 별개의 독립된 데이터로 간주하는 것입니다.

## 무엇을 기준으로 묶을까






-일단 변수들 중 텍스트로 된 데이터가 상당히 많음. 대응되는 코드가 있는 애들은 괜찮지만, 없는 애들은 텍스트마이닝을 통하여 드러나는 특정 벡터값을 근거로 수치화하여야 함. 이때 이 근거는 이미 학술적으로 증명되었고 학계에 만연한 방법론으로 따로 증명이 필요 없는 것이어야 함. 근데 아마 그렇게까지 해야 할 변수는 드물 듯. 대부분 변수들은 벡터화된 코드가 있고 없는 애들은 쓸데없이 겁나 긴 텍스트 내용물을 담고 있는 애들임.

선행연구는 애초에 사업 id가 아니라 과제를 기준으로 묶었음.
저도 과제고유번호로 묶을 것입니다.

'사업명'으로 먼저 Merge하면 안된다고 생각하는 이유:
('과제명' 으로 합치면 안된다고 보는 이유와 같습니다.)
1. '사업명' 기준 병합의 한계

후속 예산 기여분 누락: 만약 어떤 사업이 2014년에 1억 원, 2015년에 5억 원, 2016년에 10억 원의 예산을 연속해서 지원받은 뒤 2017년에 세계적인 논문을 발표했다고 가정해 보겠습니다.


모델의 잘못된 학습: 이 데이터를 단순히 '사업명'으로 묶어서 2014년 베이스라인(1억 원)에 매핑해버리면, 인공지능은 **"1억 원만 투입해도 저런 엄청난 논문이 나온다"**라고 착각하며 잘못된 패턴을 학습하게 됩니다.

결과적으로, 해당 성과를 내기 위해 투입된 진짜 동력인 후속 연구비(15억 원)의 기여분이 모델에서 완전히 무시되는 데이터 누수(Data Leakage)가 발생합니다.

2. '과제고유번호' 기준 병합의 압도적 장점

독립적인 계약 단위 식별: 과제고유번호는 다년도 사업 전체가 아니라, "해당 연도에 지급된 특정 예산 규모의 계약" 자체를 독립적으로 의미합니다.


사사(Acknowledgement) 기반의 정확한 매핑: 연구자가 논문이나 특허를 낼 때 "이 연구는 OOO년도 과제번호의 지원을 받았습니다"라고 명시해야만 성과가 연결됩니다.


깔끔한 인과관계 도출: 이를 통해 **"정확히 그 해에 투입된 그 돈(X)이 만들어낸 성과(Y)"**만을 불순물 없이 깔끔하게 발라낼 수 있습니다.

. 왜 '과제고유번호' 단위 분석이 '과제명' 단위 분석보다 ML 모델에 유리한가?   
 (사업명 기준 분석보다 유리한 이유와 동일)   
**과제명(Project Name)** 으로 데이터를 묶어버리면 머신러닝 모델이 인과관계(X $\rightarrow$ Y)를 학습할 때 심각한 오류에 빠집니다.과제명으로 묶었을 때의 오류: * A과제가 2014년에 1억, 2015년에 5억, 2016년에 10억을 받았고 2017년에 네이처에 논문을 냈다고 가정해 봅시다.이걸 과제명으로 묶어서 2014년 베이스라인(1억)의 성과로 매핑해버리면, AI는 **"아! 1억만 줘도 네이처를 쓰는구나!"**라고 잘못 학습(Data Leakage)하게 됩니다. 후속 연구비(15억)의 기여분이 무시된 것이죠.  
과제고유번호(고유계약) 단위의 깔끔함:  
과제고유번호는 "2014년에 지급된 1억짜리 계약" 자체를 의미합니다.연구자가 논문을 발표할 때, 사사(Acknowledgement)에 "이 논문은 2014년의 과제고유번호(ex. NRF-2014-001)의 지원을 받았습니다"라고 명시해야만 매핑이 됩니다.즉, **"정확히 그 해에 투입된 그 돈(X)이 만들어낸 성과(Y)"**만을 깔끔하게 발라낼 수 있습니다. 후속 연구비 통제 문제를 가장 방어하기 쉬운 논리입니다.

## 왜 선행연구가 특정 사업만 썼을까

선행연구는 '부처 간 예산 전용 불가', '균형 발전 의무' 같은 정치적/행정적 안건들을 수학적 제약조건(Constraints)으로 모두 수식화하는 것이 불가능하다고 판단했을 수 있다. 그래서 아예 "특정 사업 내에서의 배분"으로 범위를 좁혀 현실성을 확보한 것일지 모른다.  

  국가 R&D는 사업마다 목적이 너무나 다르다.

기초연구사업: 10년 뒤를 내다보고 당장의 돈(매출)이나 특허보다는 '네이처/사이언스 논문'을 쓰는 게 목적입니다.

중소기업 상용화 지원 사업: 당장 내년에 제품을 팔아서 '매출'을 올리고 '일자리'를 창출하는 게 목적입니다.
이 두 사업을 하나의 데이터셋에 섞어놓고 머신러닝을 돌리면 어떻게 될까요? 모델은 혼란에 빠집니다. **"어떤 과제는 논문이 0개인데 성공했다고 하고, 어떤 과제는 매출이 0원인데 성공했다고 하네?"**라며 학습을 포기하거나 노이즈(Noise)로 덮이게 됩니다. 사과와 오렌지를 비교하는 격이기 때문에, 선행 연구는 '동일한 목적을 가진 사과들(특정 사업)'만 모아서 분석한 것입니다.  

대학의 연구과제들 그리고 논문 + 특허만 검사하는 이유도 이러한 것일 것이다.  
만약 중소기업 과제들에 대해서 확인하는 거면 논문이나 인력양성은 주요 성과 지표가 되지 않는다. 주요 성과 지표는 기술료 또는 빠르게 수익화가 되는 것이 될 것이다.
  
내가 쓰려는 논문에서 할 일은 즉, 선행연구 대비 어떤 종류의 데이터에 대한 분석을 왜 추가했는지에 대한 설명이 필요할 것이다. 그게 아니면
BK21 사업 + 2021년 데이터 정도만 측정하고 쉽고 그다지 큰 의미는 없는 논문이 될 수도 있다. 다른 더 큰 사업이나 측정 영역을 선택하거나, 또는 다양한 집합의 사업을 한대 묶어서 분석을 진행해야 할 것이고 이러한 사업 + 산업들의 선정이 어째서 robust하고 안정적인 분석이 될 수 있는 것인지 도메인 지식을 동원해서 설명해야 할 것이다.
